```python
import pandas as pd


import warnings 
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", color_codes=True)

# загружаем файл с данными через G-colab
iris = pd.read_csv("/content/gdrive/My Drive/content/Iris.csv") # добавляем в датафрейм

# посмотрим струру данных в DF
display(iris.head())
display(iris.info()) # нет пропусков , в DF 150 строк 

```

```python
# Let's see how many examples we have of each species
iris["Species"].value_counts()
```

```python
plt.rcParams["figure.figsize"] = (10, 8) # Корректируем размер для наглядности 

iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm", marker="*",s=70)

plt.title('Comparison SepalLength/Width in cm',fontsize=16);
```

```python
# Гистограмма и диаграмма рассеивания на одном графике
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris,size=10)
```

```python

sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python

ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python

sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
```

```python

sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python

sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python

iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

```python

from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
```

```python

from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python

from pandas.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")
```
